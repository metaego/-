```python
import numpy as np
import pandas as pd
import random

def ratelist(total, rate):
    li = []

    o = 1
    x = 0

    _rate = rate
    _count = total

    for i in range(1,_count):
        if i <= _rate * _count:
            li.append(o)
        else:
            li.append(x)
    
    random.shuffle(li)
    return li
```
